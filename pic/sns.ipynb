```python
import matplotlib as plt 
# import matplotlib.pyplot as plt 
import seaborn as sns # 这个
%matplotlib inline
# 先创建画布 最后可以直接保存画布
#设置一个（12，6）的画布
plt.figure(figsize=(12, 6))
# 在jupyter中正常显示图形
```

导入数据 数据处理之缺失值

```python
titanic=sns.load_dataset('titanic')
#去除'age'中的缺失值，distplot直方图 不能处理缺失数据
age1=titanic['age'].dropna()
```

1. 直方图displot一个变量的分布
2. 条形图barplot xy分别是俩变量
3. 计数图countplot 可以调换横纵

- 参数hue设置为某种变量 可以对x轴的数据进行细分 会有自动的颜色区分   如果要对不同的类别分开绘制，用'col'参数代替'hue'
- 全是纵向的图 这里把计数图的y设置成变量会更好看

```python
sns.countplot(y='deck',hue='who',data=titanic )
```

4. 散点图stripplot/swarmplot
5. 箱线图boxplot 一般对于连续的数据范围 显示最大值、最小值、中位数及上下四分位数
6. 突然想到烛台图 开盘高低收盘 可以用matplotlib.finance
7. 小提琴图 箱线图进阶版violinplot  某个参数只有二元选择时 hue可以直接选用split参数

```python
sns.violinplot(x='class',y='age',hue='alive',
                    data=titanic,split=True)
```

8. 回归图implot 如果设置了hue那么可以markers=['*','o','+']来改变点的形状
9. 热力图heatmap 表示矩阵形状的数据 annot参数可以控制是否显示具体数值 用true或false
10. 相关系数图 这个综合一点 经常画一点 两个特征：1相关性可以自己定义2对角线通常是单一变量的直方图或核密度图

- 插播一条pivot：numpy里的函数 可以把仨数据打包成dataframe 然后才能画热力图

```python
# 例如，假设你有一个名为 'data' 的 DataFrame
data = ...  # 你的数据集

# 绘制相关系数图
correlation_matrix = data.corr()# 多计算了一个相关系数矩阵
sns.set(style="white")  # 设置样式
plt.figure(figsize=(10, 8))  # 设置图表大小
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
# 绘制热力图 其中 annot=True 表示在图中显示相关系数的数值，cmap="coolwarm"设置颜色映射，fmt=".2f"表示数值格式为两位小数。
plt.title("Correlation Matrix")  # 设置图表标题
plt.show()
```
