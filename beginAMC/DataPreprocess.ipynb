```python
import pandas as pd

# Load the uploaded Excel file
file_path = '决赛.xlsx'
df = pd.read_excel(file_path)
# Modify the "point_victor" column: change values from 2 to 0
df['point_victor'] = df['point_victor'].replace(2, 0)

# Calculate Spearman and Pearson correlation coefficients with the updated "point_victor"
spearman_corr = df.corr(method='spearman')['point_victor'].sort_values()
pearson_corr = df.corr(method='pearson')['point_victor'].sort_values()

# Combine the Spearman and Pearson correlations into a single DataFrame for comparison
correlations_combined = pd.DataFrame({
    'Spearman': spearman_corr,
    'Pearson': pearson_corr
})

correlations_combined
```

```python
# Load the Excel file again to ensure the dataframe is in its original state
df = pd.read_excel(file_path)

# 1. Delete rows where 'speed_mph', 'serve_width', 'serve_depth', or 'return_depth' is NA
df.dropna(subset=['speed_mph', 'serve_width', 'serve_depth', 'return_depth'], inplace=True)

# 2. Convert 'score' values
score_conversion = {'AD': 4, '15': 1, '30': 2, '40': 3}
df['p1_score'] = df['p1_score'].replace(score_conversion)
df['p2_score'] = df['p2_score'].replace(score_conversion)

# 3. One-hot encode 'winner_shot_type'
df['winner_shot_type_B'] = (df['winner_shot_type'] == 'B').astype(int)
df['winner_shot_type_F'] = (df['winner_shot_type'] == 'F').astype(int)

# 4. One-hot encode 'serve_width'
serve_width_columns = pd.get_dummies(df['serve_width'], prefix='serve_width')
df = pd.concat([df, serve_width_columns], axis=1)

# 5. One-hot encode 'serve_depth'
serve_depth_columns = pd.get_dummies(df['serve_depth'], prefix='serve_depth')
df = pd.concat([df, serve_depth_columns], axis=1)

# 6. One-hot encode 'return_depth'
return_depth_columns = pd.get_dummies(df['return_depth'], prefix='return_depth')
df = pd.concat([df, return_depth_columns], axis=1)

# 7. Delete original columns that were one-hot encoded
df.drop(columns=['winner_shot_type', 'serve_width', 'serve_depth', 'return_depth'], inplace=True)

# Display the modified DataFrame to confirm changes
df.head()
```
